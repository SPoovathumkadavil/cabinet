# Vectors

Vectors are the basic one-dimensional data structure in R. They are useful for small probability examples and regression inputs.

```
x <- c(2, 3, 5, 7)
x[x > 3]
```

Use this with [Linear Models](../../../statistics/regression/linear-models/index.html) and [Random Variables](../../../statistics/probability/random-variables/index.html).

# Missing Values

Missing values are represented with `NA`.

```
y <- c(1, NA, 4)
y[!is.na(y)]
```
